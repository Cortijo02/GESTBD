```python
query1 = """ 
    PREFIX sch: <https://schema.org/>
    SELECT ?schoolName ?universityName ?email ?phone ?address ?postalCode ?web
    WHERE {
      ?school a sch:School ;
            sch:legalName ?schoolName ;
            sch:containedInPlace ?university .
            OPTIONAL { ?school sch:email ?email }
            OPTIONAL { ?school sch:telephone ?phone }
            OPTIONAL { ?school sch:address ?address }
            OPTIONAL { ?school sch:url ?web }

            ?university a sch:CollegeOrUniversity ;
    }
    ORDER BY ?universityName
"""
# Asignar la consulta
sparql.setQuery(query1)
sparql.setReturnFormat(JSON)

# Ejecutar la consulta
results = sparql.query().convert()
```

```python
import requests
import json

# URL del endpoint SPARQL de GraphDB
endpoint = "http://graphdbdb:7200/repositories/Practica_GESTDB"

# Tu consulta SPARQL
query = """
PREFIX sch: <https://schema.org/>

SELECT *
WHERE { ?s ?p ?a
  }
"""

query1 = """ 
    PREFIX sch: <https://schema.org/>
    SELECT ?schoolName ?universityName ?email ?phone ?address ?postalCode ?web
    WHERE {
      ?school a sch:School ;
            sch:legalName ?schoolName ;
            sch:containedInPlace ?university .
            OPTIONAL { ?school sch:email ?email }
            OPTIONAL { ?school sch:telephone ?phone }
            OPTIONAL { ?school sch:address ?address }
            OPTIONAL { ?school sch:url ?web }

            ?university a sch:CollegeOrUniversity ;
    }
    ORDER BY ?universityName
"""

# Cabeceras adecuadas para SPARQL 1.1 POST
headers = {
    "Content-Type": "application/sparql-query",  # estándar SPARQL 1.1 para queries en POST
    "Accept": "application/sparql-results+json" # pedimos el resultado en JSON
}

# Hacer la petición POST con la consulta en el body
response = requests.post(endpoint, data=query1.encode('utf-8'), headers=headers)

# Revisar si hubo error
if response.status_code == 200:
    results = response.json()  # parsear el JSON
    for r in results["results"]["bindings"]:
        print(
            r
        )
else:
    print("Error:", response.status_code)
    print(response.text)
```
