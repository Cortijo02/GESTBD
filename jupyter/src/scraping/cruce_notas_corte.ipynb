```python
import pandas as pd
import re
```

```python
grados = pd.read_csv("grados.csv", sep=";")
grados_ofertados = pd.read_csv("grados_ofertados.csv", sep=";")
notas_corte = pd.read_csv("notas_corte.csv", sep=",")

merged = notas_corte.merge(
    grados,
    left_on="grado",
    right_on="nombre",
    how="left"
).merge(
    grados_ofertados,
    left_on="id",
    right_on="id_grado",
    how="left"
)

nota_corte = merged[["id_y", "a√±o", "convocatoria", "grupo", "nota"]].rename(
    columns={"id_y": "id_grado_ofertado"}
)

nota_corte["id_grado_ofertado"] = (
    pd.to_numeric(nota_corte["id_grado_ofertado"], errors="coerce")
    .fillna(0)
    .astype(int)
)

nota_corte.insert(0, "id", range(1, len(nota_corte) + 1))

nota_corte.to_csv("nota_corte.csv", sep=";", index=False)
```
